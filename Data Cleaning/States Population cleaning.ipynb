```python
import pandas as pd
import numpy as np
import glob
```

```python
# df = pd.concat(
#     map(pd.read_csv, ['Johor.csv', 'Kedah.csv', 'Kelantan.csv', 'Kuala Lumpur.csv', 'Labuan.csv',\
#                       'Melaka.csv', 'Negeri Sembilan.csv', 'Pahang.csv', 'Penang.csv', 'Perak.csv',\
#                       'Perlis.csv', 'Sabah.csv', 'Sarawak.csv', 'Selangor.csv', 'Terengganu.csv']), \
#     ignore_index=True)

folder_path = './Population JPM/'
csv_files = glob.glob(folder_path + '*.csv')
df = pd.concat(map(pd.read_csv, csv_files), ignore_index=True)
```

```python
df['Ethnic group'].fillna(df['Ethnic Group'], inplace=True)
df.drop('Ethnic Group', axis=1, inplace=True)
```

```python
replace_values = ['Other Malaysian', 'Other Bumiputera', 'Other Malaysian ', 'Non-Malaysian Citizens']
replacement_value = 'Others'

# Replace the values in the 'Ethnic group' column
df['Ethnic group'] = df['Ethnic group'].replace(replace_values, replacement_value)
```

```python
df['Value (\'000 Person)'] = df['Value (\'000 Person)'].replace('-', np.nan)
df['Value (\'000 Person)'] = df['Value (\'000 Person)'].astype(float)
```

```python
df[(df['State'] == 'Kelantan') & (df['Sex'] == 'Total Sex') & (df['Year'] == 1991) & (df['Age Group']=='Total Age Group')]
```

```python
mask = (df['Ethnic group'] == 'Others')

# Use the mask to get a subset dataframe for 'Others', and group by the columns to sum the values
df_grouped = df.loc[mask].groupby(['State', 'Sex', 'Age Group', 'Year', 'Ethnic group'], as_index=False)['Value (\'000 Person)'].sum()

# Remove the original rows with 'Others' from the original dataframe
df = df.loc[~mask]

# Append the summed rows to the original dataframe
df = pd.concat([df, df_grouped], ignore_index=True)
```
