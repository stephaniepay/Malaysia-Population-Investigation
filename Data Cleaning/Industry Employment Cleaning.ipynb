```python
df = pd.read_csv('2001-2021 Percentage Employed Persons by Industry and Sex.csv')
df2 = pd.read_csv('1982-2020 Employed person by sex, state.csv')
```

```python
df = df.drop('MSIC version', axis=1)
```

```python
df.head(50)
```

```python
df2 = df2[df2['State'] == 'Malaysia']
```

```python
df2.head(50)
```

```python
df_merged = pd.merge(df, df2, on=['Year', 'Sex'], how='left')
```

```python
df_merged
```

```python
df_filtered = df_merged[df_merged['Year'].between(2001, 2020)]
```

```python
df_filtered.head(50)
```

```python
df_filtered = df_filtered.rename(columns={' Employed Person in Thousands': 'Total Employed Person (\'000)'})
```

```python
df_filtered['Percentage'] = pd.to_numeric(df_filtered['Percentage'])
df_filtered['Total Employed Person (\'000)'] = pd.to_numeric(df_filtered['Total Employed Person (\'000)'])
```

```python
# df_filtered['Total Employed Person'] = df_filtered['Total Employed Person'].multiply(1000)
```

```python
df_filtered['Industry Employed Population (\'000)'] = (df_filtered['Percentage']/100) * df_filtered['Total Employed Person (\'000)']
```

```python
df_filtered['Industry Employed Population (\'000)'] = df_filtered['Industry Employed Population (\'000)'].round(1)
df_filtered
```

```python
df_filtered['Industry'].unique()
```

```python

rename_mapping = {
    'Mining and quarrying': 'Mining',
    'Manufacturing': 'Manufacturing',
    'Construction': 'Construction',
    'Public administration and defence; compulsory social security': 'Public administration',
    'Education': 'Education',
    'Agriculture, hunting and forestry': 'Agriculture',
    'Electricity, gas and water supply': 'Utilities',
    'Extra-territorial organisations and bodies': 'International organizations',
    'Financial intermediation': 'Financial',
    'Fishing': 'Fishing',
    'Health and social work': 'Healthcare',
    'Hotels and restaurants': 'Hospitality',
    'Other community, social and personal service activities': 'Services',
    'Private households with employed person': 'Households',
    'Real estate, renting and business activities': 'Real estate',
    'Transport, storage and communications ': 'Transportation',
    'Wholesale and retail trade; repair of motor vehicles, motorcycles and personal and household goods': 'Wholesale & retail trade',
    'Transportation and storage': 'Transportation',
    'Accommodation and food service activities': 'Accommodation & food services',
    'Information and communication': 'Info & communication',
    'Financial and insurance/takaful activities': 'Financial & insurance',
    'Real estate activities': 'Real estate',
    'Administrative and support service activities': 'Admin & support services',
    'Human health and social work activities': 'Healthcare',
    'Activities of households as employers': 'Households',
    'Agriculture, forestry and fishing': 'Agriculture',
    'Arts, entertainment and recreation': 'Arts & recreation',
    'Electricity, gas, steam and air conditioning supply': 'Utilities',
    'Other service activities': 'Services',
    'Professional, scientific and technical activities': 'Professional & technical activities',
    'Water supply; sewerage, waste management and remediation activities': 'Water & waste management',
    'Wholesale and retail trade, repair of motor vehicles and motorcycles': 'Wholesale & retail trade'
}

# Apply renaming using map function
df_filtered['Industry'].replace(rename_mapping, inplace=True)
df_filtered
```

```python
for year in range(2001, 2021):
    df_year = df_filtered[df_filtered['Year'] == year]
    unique_industries = df_year['Industry'].nunique()
    print(f"Year {year}: {unique_industries} unique industries")
```
