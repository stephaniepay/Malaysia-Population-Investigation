```python
import pandas as pd
import numpy as np
import glob
```

```python
df = pd.read_csv('Population 1980-2021.csv')
```

```python
df
```

```python
df = df.dropna()
```

```python
df = pd.melt(df, id_vars=['Country Name', 'Country Code'], var_name='Year', value_name='TotalPopulation')
```

```python
df = df[['Year', 'TotalPopulation']]
```

```python
df
```

```python
df['Year'] = df['Year'].str.replace(r'\[\w+\]', '', regex=True)
```

```python
df = df[~df['Year'].str.contains('Series Name|Series Code')].reset_index(drop=True)
```

```python
df.info()
```

```python
df['Year'] = df['Year'].astype(int)
df['TotalPopulation'] = df['TotalPopulation'].astype(int)
```

```python
df
```

```python
df.to_csv('filename.csv', index=False)
```
