```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./Jadual Ringkasan, 2020-2022.csv')
```

```python
df.columns = df.columns.str.strip()
df.columns
```

```python
df = df.filter(regex=r'^(?!Unnamed)')
```

```python
# df.rename(columns={'Negeri\nState': 'State'}, inplace=True)
```

```python
df = df[df['State'] != 'Total']
```

```python
years = df.columns[1:]
```

```python
melted_df = pd.melt(df, id_vars=['State'], value_vars=years, var_name='Year', value_name='Value')
```

```python
melted_df = melted_df[['State', 'Gender', 'Year', 'Value']]
```

```python
df
```

```python
df.replace('Total', np.nan, inplace=True)
df.replace('Male', np.nan, inplace=True)
df.replace('Female', np.nan, inplace=True)
df['State'].fillna(method='ffill', inplace=True)
```

```python
df['Gender'] = np.where(df.index % 18 == 0, 'Total', np.where(df.index % 18 < 18, 'Male', 'Female'))
```

```python
df_melt = pd.melt(df, id_vars=['State', 'Gender'], var_name='Year', value_name='Values')
```

```python
df_clean = df_melt.dropna(subset=['Values'])
```

```python
df_clean.reset_index(drop=True, inplace=True)
```

```python
df_clean.head(50)
df_clean.shape
```

```python
df_clean.head(50)
```

```python
# Reset the index of the DataFrame 
df_clean = df_clean.reset_index()

# Create the 'Gender' column based on the index
df_clean['Gender'] = (df_clean.index // 17) % 3
df_clean['Gender'] = df_clean['Gender'].map({0: 'Total', 1: 'Male', 2: 'Female'})

# Drop the index column that was created when resetting the index
df_clean = df_clean.drop(columns=['index'])
```

```python
df_clean.head(50)
```

```python
df_clean['Year'] = df_clean['Year'].replace('2021p', '2021')
df_clean['Year'] = df_clean['Year'].replace('2022e', '2022')
```

```python
df_clean
```

```python
df_clean['Year'] = df_clean['Year'].astype(int)
```

```python
df_clean['State'] = df_clean['State'].replace({'Sabah1': 'Sabah', 'Selangor2': 'Selangor'})
```

```python
df_clean['State'] = df_clean['State'].str.rstrip()
df_clean
```

```python
df_clean.to_csv('LifeExpectancy 2016-2022(cleaned).csv', index=False)
```
